# Python 控制流

本教程将学习Python中的控制流语句，包括条件语句和循环语句。

## 1. if 条件语句


```python
# 简单的if语句
age = 18
if age >= 18:
    print("您已经成年了")
```

```python
# if-else语句
score = 85
if score >= 60:
    print("及格了")
else:
    print("不及格")
```

```python
# if-elif-else语句
score = 85
if score >= 90:
    grade = "优秀"
elif score >= 80:
    grade = "良好"
elif score >= 70:
    grade = "中等"
elif score >= 60:
    grade = "及格"
else:
    grade = "不及格"

print(f"分数: {score}, 等级: {grade}")
```

## 2. for 循环


```python
# 基本的for循环 - 遍历列表
fruits = ["苹果", "香蕉", "橘子"]
for fruit in fruits:
    print(f"水果: {fruit}")
```

```python
# 使用range()函数
print("打印0到4:")
for i in range(5):
    print(i)

print("\n打印2到5:")
for i in range(2, 6):
    print(i)

print("\n打印0到10，步长为2:")
for i in range(0, 11, 2):
    print(i)
```

```python
# 遍历字符串
word = "Python"
for char in word:
    print(char)
```

```python
# enumerate()函数 - 同时获取索引和值
fruits = ["苹果", "香蕉", "橘子"]
for index, fruit in enumerate(fruits):
    print(f"索引 {index}: {fruit}")
```

## 3. while 循环


```python
# 基本的while循环
count = 0
while count < 5:
    print(f"计数: {count}")
    count += 1
```

```python
# while循环的实用示例 - 计算1到10的和
sum_num = 0
num = 1
while num <= 10:
    sum_num += num
    num += 1
print(f"1到10的和: {sum_num}")
```

## 4. break 和 continue


```python
# break - 跳出循环
print("使用break跳出循环:")
for i in range(10):
    if i == 5:
        break  # 当i等于5时跳出循环
    print(i)
print("循环结束")
```

```python
# continue - 跳过本次循环，继续下一次循环
print("使用continue跳过偶数:")
for i in range(10):
    if i % 2 == 0:  # 如果是偶数
        continue  # 跳过本次循环
    print(i)
```

## 5. 嵌套循环


```python
# 嵌套循环 - 打印乘法表
print("九九乘法表:")
for i in range(1, 10):
    for j in range(1, i + 1):
        print(f"{j}×{i}={i*j}", end="\t")
    print()  # 换行
```
